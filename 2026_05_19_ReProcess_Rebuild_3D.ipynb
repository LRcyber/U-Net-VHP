```python
import os
import requests

base_url = "https://data.lhncbc.nlm.nih.gov/public/Visible-Human/Male-Images/PNG_format/head/"

os.makedirs("images", exist_ok=True)

# intervalo da cabeça
for i in range(1001, 1377):
    filename = f"a_vm{i}.png"
    url = base_url + filename

    response = requests.get(url)

    if response.status_code == 200:
        with open(f"images/{filename}", "wb") as f:
            f.write(response.content)
        print(f"Baixado: {filename}")
    else:
        print(f"Erro: {filename}")

print("Download concluído")
```

```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tqdm import tqdm
```

```python
MODEL_PATH = "/content/drive/MyDrive/hvp_training/melhor_modelo_unet_512.h5"

INPUT_IMAGES_DIR = "/content/images"

OUTPUT_RGBA_DIR = "/content/png_transparente"

IMG_SIZE = 512

THRESHOLD = 0.35  # menor limiar preserva estruturas finas

os.makedirs(OUTPUT_RGBA_DIR, exist_ok=True)
```

```python
def dice_coef(y_true, y_pred, smooth=1e-6):
    y_pred = tf.cast(y_pred > 0.5, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred)
    return (2. * intersection + smooth) / (union + smooth)

def dice_loss(y_true, y_pred):
    return 1 - dice_coef(y_true, y_pred)

def combined_loss(y_true, y_pred, alpha=0.5):
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    dice = dice_loss(y_true, y_pred)
    return alpha * bce + (1 - alpha) * dice

def iou_score(y_true, y_pred):
    y_pred = tf.cast(y_pred > 0.5, tf.float32)
    intersection = tf.reduce_sum(y_true * y_pred)
    union = tf.reduce_sum(y_true) + tf.reduce_sum(y_pred) - intersection
    return (intersection + 1e-6) / (union + 1e-6)
```

```python
model = keras.models.load_model(
    MODEL_PATH,
    custom_objects={
        'combined_loss': combined_loss,
        'dice_coef': dice_coef,
        'iou_score': iou_score
    }
)

print("✅ Modelo carregado!")
```

```python
# ==========================================================
# AJUSTE DO PADDING IGUAL AO TREINAMENTO
# ==========================================================

IMG_SIZE = 512

def resize_with_padding(image, target=512):

    h, w = image.shape[:2]

    scale = min(target / w, target / h)

    nw = int(w * scale)
    nh = int(h * scale)

    resized = cv2.resize(
        image,
        (nw, nh),
        interpolation=cv2.INTER_CUBIC
    )

    canvas = np.zeros((target, target, 3), dtype=np.uint8)

    x = (target - nw) // 2
    y = (target - nh) // 2

    canvas[y:y+nh, x:x+nw] = resized

    return canvas, (x, y, nw, nh)
```

```python
# ==========================================================
# RESTAURA TAMANHO ORIGINAL
# ==========================================================

def restore_original(mask, original_shape, params):

    x, y, nw, nh = params

    crop = mask[y:y+nh, x:x+nw]

    restored = cv2.resize(
        crop,
        (original_shape[1], original_shape[0]),
        interpolation=cv2.INTER_LINEAR
    )

    return restored
```

```python
# ==========================================================
# PÓS-PROCESSAMENTO DA MÁSCARA
# ==========================================================

THRESHOLD = 0.35

def refine_mask(mask):

    mask = (mask * 255).astype(np.uint8)

    # suavização
    mask = cv2.GaussianBlur(mask, (5,5), 0)

    # threshold
    _, mask = cv2.threshold(
        mask,
        int(255 * THRESHOLD),
        255,
        cv2.THRESH_BINARY
    )

    # remove ruídos pequenos
    kernel = np.ones((3,3), np.uint8)

    mask = cv2.morphologyEx(
        mask,
        cv2.MORPH_OPEN,
        kernel,
        iterations=1
    )

    # fecha pequenos buracos
    mask = cv2.morphologyEx(
        mask,
        cv2.MORPH_CLOSE,
        kernel,
        iterations=2
    )

    return mask
```

```python
# ==========================================================
# PASTAS DE SAÍDA
# ==========================================================

OUTPUT_RGBA_DIR = "/content/png_transparente"

os.makedirs(OUTPUT_RGBA_DIR, exist_ok=True)

print("✅ Pasta criada:", OUTPUT_RGBA_DIR)
```

```python
# ==========================================================
# LISTAR IMAGENS
# ==========================================================

files = sorted([
    f for f in os.listdir(INPUT_IMAGES_DIR)
    if f.lower().endswith((".png", ".jpg", ".jpeg"))
])

print(f"📊 Total de imagens encontradas: {len(files)}")
print(files[:5])
```

```python
# ==========================================================
# PROCESSAMENTO DAS IMAGENS
# ==========================================================

from tqdm import tqdm

for file in tqdm(files):

    path = os.path.join(INPUT_IMAGES_DIR, file)

    image = cv2.imread(path)

    if image is None:
        continue

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    original_shape = image.shape[:2]

    # ------------------------------------------------------
    # RESIZE COM PADDING
    # ------------------------------------------------------

    padded, params = resize_with_padding(image)

    inp = padded.astype(np.float32) / 255.0

    inp = np.expand_dims(inp, axis=0)

    # ------------------------------------------------------
    # PREDIÇÃO
    # ------------------------------------------------------

    pred = model.predict(inp, verbose=0)[0]

    pred = pred.squeeze()

    # ------------------------------------------------------
    # RESTAURA TAMANHO ORIGINAL
    # ------------------------------------------------------

    pred = restore_original(
        pred,
        original_shape,
        params
    )

    # ------------------------------------------------------
    # REFINA MÁSCARA
    # ------------------------------------------------------

    mask = refine_mask(pred)

    # ------------------------------------------------------
    # CRIA CANAL ALPHA
    # ------------------------------------------------------

    alpha = mask.copy()

    alpha = cv2.GaussianBlur(alpha, (3,3), 0)

    # ------------------------------------------------------
    # RGBA TRANSPARENTE
    # ------------------------------------------------------

    rgba = cv2.cvtColor(image, cv2.COLOR_RGB2RGBA)

    rgba[:, :, 3] = alpha

    # remove fundo
    rgba[alpha == 0] = [0,0,0,0]

    # ------------------------------------------------------
    # SALVAR PNG TRANSPARENTE
    # ------------------------------------------------------

    output_path = os.path.join(
        OUTPUT_RGBA_DIR,
        file.replace(".jpg", ".png")
    )

    cv2.imwrite(
        output_path,
        cv2.cvtColor(rgba, cv2.COLOR_RGBA2BGRA)
    )

print("✅ Todas as imagens foram processadas!")
```

```python
# ==========================================================
# VISUALIZAÇÃO DE RESULTADO
# ==========================================================

import matplotlib.pyplot as plt

sample = files[100]

path = os.path.join(OUTPUT_RGBA_DIR, sample)

img = cv2.imread(path, cv2.IMREAD_UNCHANGED)

img = cv2.cvtColor(img, cv2.COLOR_BGRA2RGBA)

plt.figure(figsize=(10,10))
plt.imshow(img)
plt.axis("off")
plt.title(sample)
plt.show()
```
